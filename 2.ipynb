```python
import pandas as pd
```

```python
data = pd.read_csv('merged_dataset_store+train.csv')

df = pd.DataFrame(data)

df.isna().sum()
```

```python
total_cols = df.columns

print(f'Total columns: {len(total_cols)}')
print(f'Total rows: {len(df)}')
```

```python
# Loop through columns
for col in df.columns:
    if df[col].isnull().any():  # check if column has nulls
        print(f"Column: {col}, Dtype: {df[col].dtype}, Null count: {df[col].isnull().sum()}")



df_of_only_null = df[df.columns[df.isnull().any()]]

df_of_only_null
```

```python
df_of_only_null.describe()

# mean	5.430086e+03	7.222866	2008.690228	23.269093	2011.752774
```

```python
df['CompetitionDistance'].fillna(df['CompetitionDistance'].median(), inplace=True)

df['CompetitionOpenSinceMonth'].fillna(1, inplace=True)  # assume January
df['CompetitionOpenSinceYear'].fillna(
    df['CompetitionOpenSinceYear'].median(), inplace=True
)

df.loc[df['Promo2'] == 0, ['Promo2SinceWeek','Promo2SinceYear','PromoInterval']] = 0

df['Promo2SinceWeek'].fillna(0, inplace=True)
df['Promo2SinceYear'].fillna(0, inplace=True)
df['PromoInterval'].fillna(0, inplace=True)

df.isna().sum()
```

```python
df.dtypes

df['Date'] = pd.to_datetime(df['Date'])

df.dtypes

# Loop through columns and print unique values
for col in df.columns:
    unique_vals = df[col].unique()
    print(f"Column: {col}, Unique values: {unique_vals}")
```

```python
df['StateHoliday'] = df['StateHoliday'].replace({'0':0, 'a':1, 'b':2, 'c':3}).astype(int)

df['StateHoliday'].unique()

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day']   = df['Date'].dt.day
df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)
df['DayOfYear']  = df['Date'].dt.dayofyear

df
```

```python
df['CompetitionOpenDate'] = pd.to_datetime(dict(
    year=df['CompetitionOpenSinceYear'],
    month=df['CompetitionOpenSinceMonth'],
    day=1
))

df['DaysSinceCompetitionOpen'] = (df['Date'] - df['CompetitionOpenDate']).dt.days
df['DaysSinceCompetitionOpen'] = df['DaysSinceCompetitionOpen'].clip(lower=0)


def is_promo_month(row):
    if row['PromoInterval'] == 0:
        return 0
    months = row['PromoInterval'].split(',')
    month_name = row['Date'].strftime('%b')
    return 1 if month_name in months else 0

df['IsPromoMonth'] = df.apply(is_promo_month, axis=1)


df = pd.get_dummies(df, columns=['StoreType', 'Assortment'], drop_first=True)
```
